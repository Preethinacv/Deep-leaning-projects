```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

import pickle
```

```python
#load the dataset

data=pd.read_csv('ANN/Churn_Modelling.csv')
```

```python
data
```

```python
#preprocess the data
#drop irrelevenat columns

data=data.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
data
```

```python
#categorical value to encoding
label_encoder_gender=LabelEncoder()
data['Gender']=label_encoder_gender.fit_transform(data['Gender'])
data
```

```python
data['Geography'].value_counts()
```

```python
#one hot encoding
from sklearn.preprocessing import OneHotEncoder
one_hot_geo=OneHotEncoder()
geo_encoder=one_hot_geo.fit_transform(data[['Geography']])
geo_encoder
```

```python
one_hot_geo.get_feature_names_out(['Geography'])
```

```python
geo_encoder.toarray()
```

```python
geo_encoder_df=pd.DataFrame(geo_encoder.toarray(),columns=one_hot_geo.get_feature_names_out(['Geography']))
```

```python
#combine all the columns
data=pd.concat([data.drop(['Geography'],axis=1),geo_encoder_df],axis=1)
```

```python
data
```

```python
# #save the encoders and scalers
with open('label_encoder_gender.pkl','wb')as file:
    pickle.dump(label_encoder_gender, file)  

with open('one_hot_geo.pkl','wb')as file:
    pickle.dump(one_hot_geo,file)


```

```python
#didvide the data into independent and dependent features

x=data.drop('Exited',axis=1)
```

```python
y=data['Exited']

#split the data into training and testing data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

#scale the features
Scaler=StandardScaler()
x_train=Scaler.fit_transform(x_train)
x_test=Scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
with open('scaler.pkl','wb') as file:
    pickle.dump(Scaler,file)
```

```python
x_train.shape[1]
```

ANN IMPLEMENTATION

```python
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
#build our ann model

model=Sequential([
    Dense(64,activation='relu',input_shape=(x_train.shape[1],)),
    Dense(32,activation='relu'),
    Dense(1,activation='sigmoid')
])
```

```python
model.summary()
```

```python
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
loss=tensorflow.keras.losses.BinaryCrossentropy()
```

```python
#compile the model

model.compile(optimizer=opt,loss='binary_crossentropy',metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import TensorBoard,EarlyStopping

log_dir="logs/fit" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callbacks=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
#set up the early stoppung

early_stopping=EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)

```

```python
#train the model

history=model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=100,batch_size=32,
                  callbacks=[tensorflow_callbacks,early_stopping])
```

```python
model.save('model.h5')
```
